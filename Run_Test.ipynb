```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
from IPython.display import display

print(f"ğŸ“‚ Current Working Directory: {os.getcwd()}")

def show_random_test_result(experiment_name, epoch='latest', results_dir='./results'):
    """éšæœºå±•ç¤ºä¸€å¼ æµ‹è¯•ç»“æœå¯¹æ¯”å›¾"""
    img_dir = os.path.join(results_dir, experiment_name, epoch, 'images')
    if not os.path.exists(img_dir):
        print(f"âš ï¸ No results found in {img_dir}")
        return
    
    files = glob.glob(os.path.join(img_dir, '*.png'))
    if not files:
        print("ğŸ“­ No PNG files found.")
        return
        
    # é€‰ç¬¬ä¸€å¼ æˆ–è€…éšæœºä¸€å¼ 
    import random
    chosen_file = random.choice(files)
    
    img = mpimg.imread(chosen_file)
    plt.figure(figsize=(15, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"ğŸ“„ Result: {os.path.basename(chosen_file)}", fontsize=12)
    plt.show()
```
