```python
# ==========================================
# üéõÔ∏è EXPERIMENT CONTROL PANEL
# ==========================================

# 1. ÂÆûÈ™åÊ†áËØÜ
# Âª∫ËÆÆÂú®ÂêçÂ≠óÈáåÂä†‰∏ä loss Á±ªÂûãÔºåÊñπ‰æøÂå∫ÂàÜÔºå‰æãÂ¶Ç "_L1"
EXPERIMENT_NAME = "run_weight_Alldata_"  
GPU_ID = 0

# 2. Áâ©ÁêÜÂèÇÊï∞ (Z1024 Êï∞ÊçÆ)
SPACING_Z = 0.0362  
SPACING_X = 0.2     

# 3. ËÆ≠ÁªÉË∂ÖÂèÇ
BATCH_SIZE = 2
LR_G = 0.0002
LR_D_RATIO = 0.1    

# [‰øÆÊîπÁÇπ 1]ÔºöÈÖçÁΩÆÊçüÂ§±ÂáΩÊï∞Á±ªÂûãÂíåÊùÉÈáç
# -----------------------------------------------------------
PIXEL_LOSS_TYPE = 'L1'   # ÈÄâÈ°π: 'L1' (MAE) Êàñ 'L2' (MSE)
LAMBDA_PIXEL = 100.0     # ÊùÉÈáçÊé®Ëçê: L1 ÈÄöÂ∏∏ËÆæ‰∏∫ 100.0; L2 ÈÄöÂ∏∏ËÆæ‰∏∫ 1.0 Êàñ 10.0
# -----------------------------------------------------------

# 4. Â¢ûÂº∫ÂºÄÂÖ≥
USE_AUGMENTATION = False 

# 5. Ë∑ØÂæÑÈÖçÁΩÆG:\train_data\AUGAN_DATA_Nii
DATA_ROOT = "/mnt/g/train_data/AUGAN_DATA_Nii"
CHECKPOINT_ROOT = "/mnt/g/checkpoints"

# ==========================================
# ==========================================
# Ëá™Âä®ÁîüÊàêÂëΩ‰ª§Ë°å (‰øÆÂ§çÁâà - ÁßªÈô§Á©∫Ë°å)
# ==========================================
aug_flags = "" if USE_AUGMENTATION else "--no_flip --no_elastic"

# Ê≥®ÊÑèÔºö‰∏çË¶ÅÂú®Ë°å‰∏éË°å‰πãÈó¥ÁïôÁ©∫Ë°åÔºÅ
cmd = f"""
python train.py \\
    --dataroot {DATA_ROOT} \\
    --name {EXPERIMENT_NAME} \\
    --model augan \\
    --netG unet_3d \\
    --batch_size {BATCH_SIZE} \\
    --pixel_loss {PIXEL_LOSS_TYPE} \\
    --lambda_pixel {LAMBDA_PIXEL} \\
    --lr {LR_G} \\
    --lr_d_ratio {LR_D_RATIO} \\
    --spacing_z {SPACING_Z} \\
    --spacing_x {SPACING_X} \\
    {aug_flags} \\
    --gpu_ids {GPU_ID} \\
    --checkpoints_dir {CHECKPOINT_ROOT} \\
    --num_workers 16 
"""

print("üöÄ Ready to launch command:")
print(cmd)
```
