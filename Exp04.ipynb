```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from IPython.display import display, clear_output
import glob

# è®¾ç½®å·¥ä½œç›®å½• (ç¡®ä¿æŒ‡å‘é¡¹ç›®æ ¹ç›®å½•)
# å¦‚æœæ‚¨çš„ notebook åœ¨é¡¹ç›®æ ¹ç›®å½•ä¸‹ï¼Œå°±ä¸éœ€è¦æ”¹ã€‚
# os.chdir('/home/liujia/AUGAN_3D_Recon') 
print(f"ğŸ“‚ Current Working Directory: {os.getcwd()}")

def show_latest_visuals(experiment_name, output_root='./project_assets/experiments'):
    """
    è‡ªåŠ¨æŸ¥æ‰¾å¹¶æ˜¾ç¤ºæœ€æ–°çš„è®­ç»ƒé‡‡æ ·å›¾ã€‚
    """
    img_dir = os.path.join(output_root, experiment_name, 'web_images')
    
    if not os.path.exists(img_dir):
        print(f"âš ï¸ No images found in {img_dir}. Training might not have started or saved yet.")
        return

    # æ‰¾æœ€æ–°çš„ png
    list_of_files = glob.glob(os.path.join(img_dir, '*.png'))
    if not list_of_files:
        print("ğŸ“­ No PNG files found yet.")
        return
        
    latest_file = max(list_of_files, key=os.path.getctime)
    
    # æ˜¾ç¤º
    img = mpimg.imread(latest_file)
    plt.figure(figsize=(15, 6)) # å®½ä¸€ç‚¹ï¼Œé€‚åº”æ¨ªå‘å›¾
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"ğŸ“¸ Latest Visual: {os.path.basename(latest_file)}", fontsize=12)
    plt.show()
```

```python
# ==========================================
# ğŸ›ï¸ EXPERIMENT CONTROL PANEL
# ==========================================

# 1. å®éªŒæ ‡è¯†
# å»ºè®®åœ¨åå­—é‡ŒåŠ ä¸Š loss ç±»å‹ï¼Œæ–¹ä¾¿åŒºåˆ†ï¼Œä¾‹å¦‚ "_L1"
EXPERIMENT_NAME = "run_04_z1024_L1_loss"  
GPU_ID = 0

# 2. ç‰©ç†å‚æ•° (Z1024 æ•°æ®)
SPACING_Z = 0.0362  
SPACING_X = 0.2     

# 3. è®­ç»ƒè¶…å‚
BATCH_SIZE = 2
LR_G = 0.0002
LR_D_RATIO = 0.1    

# [ä¿®æ”¹ç‚¹ 1]ï¼šé…ç½®æŸå¤±å‡½æ•°ç±»å‹å’Œæƒé‡
# -----------------------------------------------------------
PIXEL_LOSS_TYPE = 'L1'   # é€‰é¡¹: 'L1' (MAE) æˆ– 'L2' (MSE)
LAMBDA_PIXEL = 100.0     # æƒé‡æ¨è: L1 é€šå¸¸è®¾ä¸º 100.0; L2 é€šå¸¸è®¾ä¸º 1.0 æˆ– 10.0
# -----------------------------------------------------------

# 4. å¢å¼ºå¼€å…³
USE_AUGMENTATION = False 

# 5. è·¯å¾„é…ç½®
DATA_ROOT = "/mnt/g/1024"
CHECKPOINT_ROOT = "/mnt/g/checkpoints"

# ==========================================
# ==========================================
# è‡ªåŠ¨ç”Ÿæˆå‘½ä»¤è¡Œ (ä¿®å¤ç‰ˆ - ç§»é™¤ç©ºè¡Œ)
# ==========================================
aug_flags = "" if USE_AUGMENTATION else "--no_flip --no_elastic"

# æ³¨æ„ï¼šä¸è¦åœ¨è¡Œä¸è¡Œä¹‹é—´ç•™ç©ºè¡Œï¼
cmd = f"""
python train.py \\
    --dataroot {DATA_ROOT} \\
    --name {EXPERIMENT_NAME} \\
    --model augan \\
    --netG unet_3d \\
    --batch_size {BATCH_SIZE} \\
    --pixel_loss {PIXEL_LOSS_TYPE} \\
    --lambda_pixel {LAMBDA_PIXEL} \\
    --lr {LR_G} \\
    --lr_d_ratio {LR_D_RATIO} \\
    --spacing_z {SPACING_Z} \\
    --spacing_x {SPACING_X} \\
    {aug_flags} \\
    --gpu_ids {GPU_ID} \\
    --checkpoints_dir {CHECKPOINT_ROOT} \\
    --num_workers 16 
"""

print("ğŸš€ Ready to launch command:")
print(cmd)
```
